```python
col = df.select_dtypes('object').columns
for j in range(1,len(col)):
        df[col[j]] = df[col[j]].str.replace("%" , "")

for i in range(1,len(col)):
    df[col[i]] = pd.to_numeric(df[col[i]], errors='coerce')
```

```python
col_1 = df.select_dtypes('float64').columns
for k in range(len(col_1)):
    df[col_1[k]] = df[col_1[k]] / 100
```

```python
df.info()
```

```python
df = df.drop(columns = ['CoC Number','Total Year-Round Beds (ES).1','Total Year-Round Beds (TH).1','Total Year-Round Beds (SH).1','HMIS Participation Rate for Year-Round Beds (SH)', 'HMIS Participation Rate for Year-Round Beds (OPH)'])
```

```python
df.describe()
```

```python
droplist = [col for col in df.columns if (df[col]== 0).sum() > 75]
```

```python
droplist
```

```python
df.drop(droplist,axis=1,inplace=True)
```

```python
df = df.fillna(df.mean())
```

```python
df = df.drop(axis = 0, index = 391)
```

```python
df.head()
```

```python
df.shape
```

```python
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features 
df.drop(to_drop, axis=1, inplace=True)
```
