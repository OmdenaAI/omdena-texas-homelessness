```python
df['Year'] = pd. to_datetime(df['Year'])
```

```python
df['Year']= pd.DatetimeIndex(df['Year']).year
```

```python
df.head()
```

```python
df = df.drop(columns =['State','CoC Name'])
```

```python
df['Count'] = df['Count'].str.replace(',', '')
```

```python
df['Count'] = df['Count'].astype('int64')
```

```python
df.isnull().sum()
```

```python
df['Measures'].value_counts()
```

```python
df.shape
```

```python
df['Index'] = df['Measures']
```

```python
df_1 = df.pivot_table('Count', ['Year','CoC Number'], 'Index')
```

```python
df_1
```

```python
df_1.reset_index(drop = False, inplace = True)
```

```python
df_1.isnull().sum()
```

```python
df_1 = df_1.dropna(axis = 1, thresh = 77, inplace= False)
```

```python
df_1.head()
```

```python
df_1.info()
```

```python
def percentage(x):
    for columns in x.columns[3:]:
        x[columns + '(%)'] = x[columns] / x['Total Homeless']
    return x
```
